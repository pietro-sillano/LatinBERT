```python
import pickle
import os
import numpy as np
import umap
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
source='/home/wsojka00/Desktop/pickle'
```

```python
def get_author(filename):
    pos1 = filename.find('_')
    return filename[:pos1]

author=[]
for f in os.listdir(source):
    author.append(get_author(f))
len(author)
```

```python
token=[]
author=[]
source='/home/wsojka00/Desktop/pickle'


for file in os.listdir(source):
    dbfile = open(source+'/'+file,'rb')     
    db = pickle.load(dbfile)
    tmp=get_author(file)
    for key in db.keys():
        token.append(db[key][2])
        author.append(tmp)    
token=np.array(token)
print(token.shape,len(author))
```

```python
np.unique(author)
```

```python
tokenvoragine = []
for i in range(len(token)):
    if author[i]=='voragine':
        tokenvoragine.append(token[i])

tokenvergil = []
for i in range(len(token)):
    if author[i]=='vergil':
        tokenvergil.append(token[i])
```

```python
tokenvoragine = [tok for tok in token if tok==0]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
label=le.fit_transform(author)
label.shape
```

```python
token,label,author=shuffle(token,label,author,random_state=0)
```

```python
import umap.plot
mapper = umap.UMAP(n_components=2,metric='euclidean', n_neighbors=50).fit(token)
```

```python
neighbours=[5,10,20,30,50,100]

for neighbour in neighbours:
    mapper = umap.UMAP(n_components=2,metric='euclidean', n_neighbors=neighbour).fit(token)
    umap.plot.points(mapper, labels=np.array(author))
```

```python
authordict = dict()

for key in np.unique(author):
    authordict[key]=[]
    
for i in range(len(token)):
    authordict[author[i]].append(token[i])
```
